```python
import random
import ollama


def gerar_dados_simulados():
    return {
        "temperatura": random.randint(-10, 45),
        "energia": random.randint(0, 100),
        "comunicacao": random.randint(0, 100),
        "oxigenio": random.randint(40, 100),
        "estabilidade": random.randint(0, 100)
    }


def analisar_dados_com_ia(dados):
    prompt_sistema = """Você é a IA de Contingência do Módulo S.E.E.D. na Missão Artemis.
    A sua função é analisar os dados vitais da estufa espacial.
    Seja extremamente direto, aja como um computador de bordo e sugira 3 ações de emergência em formato de tópicos (bullet points)."""

    prompt_usuario = f"""
    DADOS DE TELEMETRIA ATUAIS:
    Temperatura: {dados['temperatura']}°C
    Energia: {dados['energia']}%
    Comunicação: {dados['comunicacao']}%
    Oxigênio: {dados['oxigenio']}%
    Estabilidade: {dados['estabilidade']}%
    Avalie o nível de risco e indique as ações de contenção.
    """

    resposta = ollama.chat(
        model="llama3.2:1b",
        messages=[
            {"role": "system", "content": prompt_sistema},
            {"role": "user", "content": prompt_usuario}
        ]
    )
    return resposta["message"]["content"]


def exibir_status(dados, analise_ia):
    print("MISSION CONTROL - STATUS ATUAL (MÓDULO S.E.E.D.)\n")
    print(f"Temperatura: {dados['temperatura']} °C")
    print(f"Energia: {dados['energia']} %")
    print(f"Comunicação: {dados['comunicacao']} %")
    print(f"Oxigênio: {dados['oxigenio']} %")
    print(f"Estabilidade: {dados['estabilidade']} %\n")

    if dados['energia'] < 20 or dados['temperatura'] > 35:
        print("ALERTA CRÍTICO: parâmetros vitais fora da margem de segurança.")
        print("Ação automática: ativar modo de economia e resfriamento.\n")
    else:
        print("Sistemas operando dentro da normalidade.\n")

    print("Análise preditiva da IA:")
    print(analise_ia)


print("Sincronizando com sensores do S.E.E.D... aguarde a análise da IA.\n")
dados_sensores = gerar_dados_simulados()
analise_gerada = analisar_dados_com_ia(dados_sensores)
exibir_status(dados_sensores, analise_gerada)
```
